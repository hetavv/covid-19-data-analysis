```python
df = pd.read_csv('Covid-19-Preprocessed-Dataset/preprocessed/covid_19_data_cleaned.csv', parse_dates=['Date'])
country_daywise = pd.read_csv('Covid-19-Preprocessed-Dataset/preprocessed/country_daywise.csv', parse_dates=['Date'])
coutnrywise = pd.read_csv('Covid-19-Preprocessed-Dataset/preprocessed/countrywise.csv')
daywise = pd.read_csv('Covid-19-Preprocessed-Dataset/preprocessed/daywise.csv', parse_dates=['Date'])
```

```python
#filling empty provience and state values as empty strings
df['Province/State'] = df['Province/State'].fillna("")
df
```

```python
confirmed = df.groupby('Date').sum()['Confirmed'].reset_index()
recovered = df.groupby('Date').sum()['Recovered'].reset_index()
deaths = df.groupby('Date').sum()['Deaths'].reset_index()
confirmed
```

```python
#plot the latest confirmed, recovered and deaths on a scatter plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=confirmed['Date'], y= confirmed['Confirmed'], name='Confirmed'))
fig.add_trace(go.Scatter(x=recovered['Date'], y= recovered['Recovered'], name ='Recovered', line=dict(color="Green")))
fig.add_trace(go.Scatter(x=deaths['Date'], y= deaths['Deaths'], name='deaths', line=dict(color="Red")))
fig.update_layout(title = 'Worldwide Covid-19 Cases', xaxis_tickfont_size = 12, yaxis = dict(title = 'Number of World wide Cases'))
fig.show()
```
